```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

## Hopper-v5

```python
file_name = "Hopper-v5-DDPG.csv"
data = pd.read_csv(file_name)
data.head()
```

```python
fig, ax = plt.subplots()
ax.plot(data["timestep"], data["mean_episode_return"])
ax.set_xlabel("Timesteps")
ax.set_ylabel("Mean episode return")
plt.show()
```
