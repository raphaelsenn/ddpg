```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

## HalfCheetah-v5

```python
file_name = "HalfCheetah-v5-DDPG.csv"
data = pd.read_csv(file_name)
data.head()
```

```python
t = data["timestep"]
mean = data["mean_episode_return"]
std = data["std_episode_return"]
lower = mean - std
upper = mean + std

fig, ax = plt.subplots()
ax.plot(t, mean, color="red", alpha=0.8)
ax.fill_between(t, lower, upper, alpha=0.2, color="red")
ax.set_xlabel("Time steps (1e-6)")
ax.set_ylabel("Average Return")
plt.show()
```
